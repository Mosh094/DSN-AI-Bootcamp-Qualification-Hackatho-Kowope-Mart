# Convert all data to numeric

```python
df_test.nunique() #For test data
```

This will not be done on "Applicant_ID" alone as it will be dropped because all entries for the column is unique, 
so no effect on the prediction result.

Import LabelEncoder library 

```python
from sklearn.preprocessing import LabelEncoder
```

Convert categorical('form_field47', 'default_status') and  data to numeric using LabelEncoding

```python
#For Train Dataset
le = LabelEncoder()
df_train["form_field47"] = le.fit_transform(df_train["form_field47"].astype(str))
df_train["default_status"] = le.fit_transform(df_train["default_status"].astype(str))
df_train
```

```python
#For Test Dataset
le = LabelEncoder()
df_test["form_field47"] = le.fit_transform(df_test["form_field47"].astype(str))
df_test
```

# Build Model


Drop 'Applicant_ID' as stated earlier

```python
X_train = df_train.drop(["Applicant_ID", "default_status"], axis =1)
y_train = df_train["default_status"]

X_test = df_test.drop(["Applicant_ID"], axis = 1)
```

Import Machine Learning Packages

```python
# ML Pkg
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

Split Training and Test Data

```python
x_train,x_validation,y_train,y_validation = train_test_split(X_train,y_train,test_size=0.30,random_state=7)
```

```python
classifier=RandomForestClassifier()
classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import roc_auc_score
```

Model Accuracy

```python
prediction = classifier.predict(x_validation)
```

```python
score = roc_auc_score(y_validation, prediction)
```

```python
print(f"Area Under the Curve: {score}.")
```

Making Predictions on Test Dataset

```python
Test_prediction = classifier.predict_proba(X_test)[:,1] #predicting using test dataset.  showing the Applicant_ID column and the prediction alone
```

```python
pd.DataFrame({"Applicant_ID": df_test["Applicant_ID"], "default_status": Test_prediction}).to_csv("Submission.csv", index = False)
```
